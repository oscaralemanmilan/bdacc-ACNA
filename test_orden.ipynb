```python
import pandas as pd

# 1. Carrega el teu fitxer (canvia el nom si cal)
file_path = "data/bd_accidents_200726_net_c.xlsx"  # El teu fitxer Excel
df = pd.read_excel(file_path)

print("--- DIAGNÒSTIC INICIAL ---")
print(f"Tipus de dada de la columna 'Data': {df['Data'].dtype}")
print("Primeres 5 files sense tocar res:")
print(df['Data'].head())

# 2. NETEJA AGRESSIVA
# Convertim a string, treiem espais i eliminem caràcters invisibles
df['Data_Neta'] = df['Data'].astype(str).str.strip()

# 3. CONVERSIÓ FORÇOSA
# dayfirst=True és CRÍTIC per a formats DD/MM/YYYY
df['Data_Python'] = pd.to_datetime(
    df['Data_Neta'], 
    dayfirst=True, 
    errors='coerce'
)

# 4. ORDENAMENT
# 'ascending=False' per veure el 2026 a dalt
df_ordenat = df.sort_values(by='Data_Python', ascending=False)

print("\n--- RESULTAT DE L'ORDENAMENT ---")
# Mostrem la columna original i la nova per comparar
columnes_interes = ['Data', 'Data_Python', 'Lloc']
print(df_ordenat[columnes_interes].head(15))

# 5. COMPROVACIÓ DE FALLADES
errors = df_ordenat[df_ordenat['Data_Python'].isna()]
if not errors.empty:
    print("\n⚠️ ALERTA: S'han trobat dates que no s'han pogut convertir:")
    print(errors['Data'])

# Opcional: Guarda el resultat per revisar-lo en Excel
# df_ordenat.to_excel("revisio_dates.xlsx", index=False)
```
